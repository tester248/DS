## Data Formatting

### 1. Checking Data Types (`df.dtypes`)

```python
# Display the current data types of the DataFrame
print("Current Data Types:")
display(df.dtypes)
```

### 2. Converting Data Types (`df.astype()` and `pd.to_datetime`)

```python
# Convert 'Request timestamp' and 'Drop timestamp' to datetime objects
# Using errors='coerce' will turn unparseable dates into NaT (Not a Time)
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'], errors='coerce')
df['Drop timestamp'] = pd.to_datetime(df['Drop timestamp'], errors='coerce')

# Convert 'Driver id' to integer type, handling NaN values by allowing them (float still works for NaN)
df['Driver id'] = df['Driver id'].astype('Int64')

print("\nData Types after conversion:")
display(df.dtypes)
display(df.head())
```

## Encoding Categorical Features

### 1. Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to the 'Pickup point' column
df['Pickup point_encoded'] = le.fit_transform(df['Pickup point'])

print("\nDataFrame with Label Encoded 'Pickup point' column:")
display(df[['Pickup point', 'Pickup point_encoded']].head())
```

### 2. One-Hot Encoding

```python
# Apply One-Hot Encoding to the 'Status' column

df_one_hot = pd.get_dummies(df, columns=['Status'], prefix='Status')

print("\nDataFrame with One-Hot Encoded 'Status' column (first 5 rows and relevant columns):")
display(df_one_hot[['Status_Cancelled', 'Status_No Cars Available', 'Status_Trip Completed']].head())
```

##Pre-Processing Steps

1. Normalization

```python
from sklearn import preprocessing
print("Raw Data:")
print(df.head())

# Normalization only works on numbers.
x = df.select_dtypes(include=['number'])

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
# This calculates the min/max and scales the data in one go
x_scaled = min_max_scaler.fit_transform(x)

# Run the normalizer on the dataframe
# We turn the resulting array back into a readable Pandas DataFrame
df_normalized = pd.DataFrame(x_scaled, columns=x.columns)

print("\nNormalized Data (Values scaled between 0 and 1):")
print(df_normalized.head())
```
