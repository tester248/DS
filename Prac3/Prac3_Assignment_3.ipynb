## Part 1 — Titanic dataset (grouped summary)

```python
# titanic = pd.read_csv('Titanic-Dataset.csv')
# titanic.head()
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

```python
# Group `Fare` by `Sex` and compute summary statistics
grouped = titanic.groupby('sex')['fare']
summary = grouped.agg(['mean','median','min','max','std'])
summary
```

```python
# Create a list of Fare values per Sex group
fare_lists = titanic.groupby('sex')['fare'].apply(list)
# display the count and first 10 values for each group
{k: {'count': len(v), 'first_10': v[:10]} for k,v in fare_lists.items()}
```

## Part 2 — Iris dataset (descriptive statistics by species)

```python
# iris = pd.read_csv('Iris.csv')
# iris.head()
iris = sns.load_dataset("iris")
iris.head()
```

```python
if 'species' in iris.columns:
    species_col = 'species'
elif 'Species' in iris.columns:
    species_col = 'Species'
else:
    raise KeyError('No species column found in Iris.csv')
grouped_iris = iris.groupby(species_col)
# Mean and std per species
means = grouped_iris.mean()
stds = grouped_iris.std()
means, stds
```

```python
# Full descriptive statistics (count, mean, std, min, 25%, 50%, 75%, max) per species
grouped_iris.describe().transpose()
```
