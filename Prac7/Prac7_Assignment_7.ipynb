**Tokenization, POS tagging, Stemming, and Lemmatization**


```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

text = 'studies studying cries cry Jupiter is the largest Planet'
# Tokenize
words = word_tokenize(text)
sentences = sent_tokenize(text)
print('Sentences:', sentences)
print('Words:', words)

# Stemming with PorterStemmer
ps = PorterStemmer()
stems = [ps.stem(w) for w in words]
print('Stems:', stems)

# Lemmatization (no POS)
lemmatizer = WordNetLemmatizer()
lemmas_no_pos = [lemmatizer.lemmatize(w) for w in words]
print('Lemmas (no POS):', lemmas_no_pos)

# POS tagging and lemmatization with POS mapping
pos_tags = pos_tag(words)
print('POS tags:', pos_tags)

def _map_pos(tag):
    if tag.startswith('J'):
        return 'a'
    elif tag.startswith('V'):
        return 'v'
    elif tag.startswith('N'):
        return 'n'
    elif tag.startswith('R'):
        return 'r'
    else:
        return 'n'

lemmas_with_pos = [lemmatizer.lemmatize(w, _map_pos(t)) for w,t in pos_tags]
print('Lemmas (with POS):', lemmas_with_pos)
```

```python
# TF-IDF demonstration: stopwords removal, TF, IDF, and TF-IDF
import nltk
nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Sample corpus
corpus = [
    'Studies studying cries cry Jupiter is the largest planet',
    'The study of planets and Jupiter shows that Jupiter is very large and gaseous'
]

# Preprocess: lowercase, tokenise, remove stopwords
stop_words = set(stopwords.words('english'))
def preprocess(doc):
    tokens = [w.lower() for w in word_tokenize(doc) if w.isalpha()]
    return ' '.join([w for w in tokens if w not in stop_words])

proc_corpus = [preprocess(d) for d in corpus]
print('Preprocessed corpus:', proc_corpus)

# Term Frequency (CountVectorizer) and normalized TF
cv = CountVectorizer()
tf = cv.fit_transform(proc_corpus).toarray()
terms = cv.get_feature_names_out()
tf_norm = tf / tf.sum(axis=1, keepdims=True)
print('\nTerms:', list(terms))
print('\nRaw TF matrix:\n', tf)
print('\nNormalized TF matrix:\n', tf_norm)

# TF-IDF using scikit-learn
tfidf = TfidfVectorizer()
tfidf_mat = tfidf.fit_transform(proc_corpus).toarray()
tfidf_terms = tfidf.get_feature_names_out()
print('\nTF-IDF terms:', list(tfidf_terms))
print('\nTF-IDF matrix:\n', tfidf_mat)
```
