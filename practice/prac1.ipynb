```python
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'], errors='coerce')
df['Drop timestamp'] = pd.to_datetime(df['Drop timestamp'], errors = 'coerce')

df['Driver id'] = df['Driver id'].fillna(0).astype(int)


print(df.dtypes)

print("Head")
display(df.head)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Pickup point Encoded '] = le.fit_transform(df['Pickup point'])
df['Status Encoded'] = le.fit_transform(df['Status'])

print("DataFrane with Encoded values")

display(df.head)
```

```python
df_one_hot = pd.get_dummies(df, columns=['Pickup point', 'Status'])
print("DataFrame with One Hot Encoded values")
display(df_one_hot.head)
```

```python
from sklearn.preprocessing import MinMaxScaler
x = df.select_dtypes(include=['number'])
print("Numerical columns")
display(x.head)

minmax_scalar = MinMaxScaler()
x_scaled = minmax_scalar.fit_transform(x)
df_normalized = pd.DataFrame(x_scaled, columns=x.columns)
print("Normalized DataFrame")
display(df_normalized.head)
```
