```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
titanic = sns.load_dataset('titanic')
print(titanic.head())
print(titanic.info())
```

```python
grouped = titanic.groupby('who')['age']
summary = grouped.agg(['mean', 'median', 'min', 'max', 'std', 'count'])
summary
```

```python
fare_lists = titanic.groupby('class')['fare'].apply(list)
print(fare_lists)
```

```python
iris = sns.load_dataset('iris')
print(iris.head())
print(iris.describe())
print(iris.info())
```
