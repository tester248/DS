# Practical 8 — Data Visualization I (Lean)
Core plotting examples using the Titanic dataset (heatmap, boxplot, distribution).

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_style('whitegrid')
ds = sns.load_dataset('titanic')
ds.head()
```

```python
# Correlation heatmap (numeric columns)
num = ds.select_dtypes(include=['number'])
corr = num.corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
```

```python
# Fare by class and age distribution
plt.figure(figsize=(8,4))
sns.boxplot(x='class', y='fare', data=ds)
plt.title('Fare by Class')
plt.show()

plt.figure(figsize=(8,4))
sns.histplot(ds['age'].dropna(), kde=True, bins=30)
plt.title('Age distribution')
plt.show()
```
